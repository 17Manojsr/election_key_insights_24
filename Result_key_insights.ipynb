```python
# Ensure 'Margin' is numeric
input_data['Margin'] = pd.to_numeric(input_data['Margin'], errors='coerce')

# Group by 'Leading Party' and sum 'Margin'
party_votes = input_data.groupby('Leading Party')['Margin'].sum().sort_values(ascending=False)

# The party with the highest and lowest margin of victory
highest_margin = input_data.loc[input_data['Margin'].idxmax()]
lowest_margin = input_data.loc[input_data['Margin'].idxmin()]
```

```python
# Number of seats won by each party
seats_won = input_data['Leading Party'].value_counts()

custom_palette = ['#FF5733', '#33FF57', '#3357FF', '#FF33A1', '#A133FF', '#33FFA1']

# Plot number of seats won by each party
plt.figure(figsize=(20, 8))
sns.barplot(x=seats_won.index, y=seats_won.values, palette=custom_palette)
plt.title('Number of Seats Won by Each Party')
plt.xlabel('Party')
plt.ylabel('Seats Won')
plt.xticks(rotation=90)
plt.show()

# Results of the highest and lowest margin victory
print("Party with the highest margin of victory:")
print(highest_margin)

print("\nParty with the lowest margin of victory:")
print(lowest_margin)
```

```python
rahul_entries = input_data[input_data['Leading Candidate'] == 'RAHUL GANDHI']
modi_entries = input_data[input_data['Leading Candidate'] == 'NARENDRA MODI']
amit_entries = input_data[input_data['Leading Candidate'] == 'AMIT SHAH']

# Get the votes for Rahul Gandhi, Narendra Modi, and Amit Shah
rahul_votes = rahul_entries['Margin'].values
modi_votes = modi_entries['Margin'].values[0] if not modi_entries.empty else 0
amit_votes = amit_entries['Margin'].values[0] if not amit_entries.empty else 0
```

```python
# Get the original constituency names for Rahul Gandhi
rahul_constituencies = list(rahul_entries['Constituency'])

# Get the original constituency name for Narendra Modi
modi_constituency = modi_entries['Constituency'].values[0] if not modi_entries.empty else "Modi Constituency"

# Get the original constituency name for Amit Shah
amit_constituency = amit_entries['Constituency'].values[0] if not amit_entries.empty else "Amit Shah Constituency"

# Combine the data
data_to_plot = pd.DataFrame({
    'Candidate': ['Rahul Gandhi'] * len(rahul_votes) + ['Narendra Modi', 'Amit Shah'],
    'Constituency': rahul_constituencies + [modi_constituency, amit_constituency],
    'Votes': list(rahul_votes) + [modi_votes, amit_votes]
})

# Plot the comparison
plt.figure(figsize=(12, 6))
sns.barplot(data=data_to_plot, x='Constituency', y='Votes', hue='Candidate', palette='muted')
plt.title('Comparison of Votes for Rahul Gandhi, Narendra Modi, and Amit Shah')
plt.xlabel('Constituency')
plt.ylabel('Votes')
plt.xticks(rotation=45)
plt.show()
```

```python
highest_margin_entry = input_data.loc[input_data['Margin'].idxmax()]
lowest_margin_entry = input_data.loc[input_data['Margin'].idxmin()]

# Combine the data
data_to_plot = pd.DataFrame({
    'Candidate': [highest_margin_entry['Leading Candidate'], lowest_margin_entry['Leading Candidate']],
    'Party': [highest_margin_entry['Leading Party'], lowest_margin_entry['Leading Party']],
    'Margin': [highest_margin_entry['Margin'], lowest_margin_entry['Margin']]
})

custom_palette = ['#FF5733', '#33FF57']
# Plot the comparison
plt.figure(figsize=(10, 6))
sns.barplot(data=data_to_plot, x='Candidate', y='Margin', hue='Party', palette=custom_palette)
plt.title('Candidates with Highest and Lowest Margin of Victory')
plt.xlabel('Candidate')
plt.ylabel('Margin of Victory')
plt.xticks(rotation=45)
plt.show()
```

```python

input_data['Margin'] = pd.to_numeric(input_data['Margin'], errors='coerce')

# colors for histogram and KDE line
hist_color = 'green'
kde_color = '#E75480'

# To Plot the histogram with KDE
plt.figure(figsize=(10, 6))
sns.histplot(input_data['Margin'], bins=20, kde=True, color=hist_color)
sns.kdeplot(input_data['Margin'], color=kde_color)
plt.title('Histogram of Margin of Victory')
plt.xlabel('Margin of Victory')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.colors as mcolors
from mpl_toolkits.axes_grid1 import make_axes_locatable

# 3.Load election data
input_data = pd.read_csv('/content/drive/MyDrive/RA2111027020103/Election_results_2024.csv')

# Inspect column names and data
print("Columns in input_data:", input_data.columns)
print(input_data.head())

# Load geographic data
geo_data = gpd.read_file('/content/drive/MyDrive/RA2111027020103/india_ds.shp')

# Inspect column names and data
print("Columns in geo_data:", geo_data.columns)
print(geo_data.head())
```

```python
# 4. Geographic distribution
input_data['Constituency'] = input_data['Constituency'].str.upper()
geo_data['DISTRICT'] = geo_data['DISTRICT'].str.upper()

# Replace non-numeric 'Margin' values with NaN and then drop them
input_data['Margin'] = pd.to_numeric(input_data['Margin'], errors='coerce')
input_data.dropna(subset=['Margin'], inplace=True)

# Merge the data
merged_data = pd.merge(geo_data, input_data, how='inner', left_on='DISTRICT', right_on='Constituency')

# Increase the size of the plot
fig, ax = plt.subplots(1, 1, figsize=(20, 15))  # Adjust the width and height as needed

# Define your custom colors for the colormap
custom_colors = ['#FFFFCC', '#A1DAB4', '#41B6C4', '#2C7FB8', '#253494']

# Create a custom colormap
custom_cmap = mcolors.LinearSegmentedColormap.from_list("custom", custom_colors)

# Plot the merged data
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)

# Plot with the custom colormap
merged_data.plot(column='Margin', cmap=custom_cmap, linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, cax=cax)
ax.set_title('Geographic Distribution of Election Margins', fontsize=20)
ax.set_xlabel('Longitude', fontsize=14)
ax.set_ylabel('Latitude', fontsize=14)

# Adjust the legend
norm = mcolors.Normalize(vmin=merged_data['Margin'].min(), vmax=merged_data['Margin'].max())
sm = plt.cm.ScalarMappable(cmap=custom_cmap, norm=norm)
sm.set_array([])

# Add a long legend outside the main plot
legend = ax.figure.colorbar(sm, ax=ax, orientation='vertical')
legend.set_label('Margin', fontsize=14)

# Save the plot to a file
plt.savefig('/content/drive/MyDrive/RA2111027020103/election_margins_custom_colormap.png', bbox_inches='tight')

# Display the plot
plt.show()
```

```python
# 5. Party Performance
party_performance = input_data['Leading Party'].value_counts().nlargest(5)

# Plotting the graph
plt.figure(figsize=(10, 6))
party_performance.plot(kind='bar', color='skyblue')
plt.title('Top 5 Parties by Seats')
plt.xlabel('Party')
plt.ylabel('Number of Seats')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6. Close Contests
close_contests = input_data[input_data['Margin'] < 1000].shape[0]

# Plotting the graph with custom colors
plt.figure(figsize=(8, 6))
sns.countplot(x='Close Contests', data=pd.DataFrame({'Close Contests': ['< 1000', '>= 1000'], 'Count': [close_contests, len(input_data) - close_contests]}), palette=['#1f77b4', '#ff7f0e'])
plt.title('Number of Close Contests (Margin < 1000)')
plt.xlabel('Margin of Victory')
plt.ylabel('Number of Contests')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 7. Incumbency Advantage
incumbency_advantage = input_data[input_data['Status'] == 'Result Declared']['Leading Party'].value_counts().nlargest(3)

# Plotting the graph with custom colors
plt.figure(figsize=(8, 6))
incumbency_advantage.plot(kind='bar', color=['#1f77b4', '#ff7f0e', '#2ca02c'])
plt.title('Top 3 Parties with Incumbency Advantage')
plt.xlabel('Party')
plt.ylabel('Number of Seats')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 8. Party Fragmentation
party_counts = input_data['Leading Party'].value_counts()
top_n = 10  # Change to show top n parties

# Get top n parties and their counts
top_parties = party_counts.nlargest(top_n)
other_parties_count = party_counts.nsmallest(len(party_counts) - top_n).sum()

# Combine top n parties and others
top_parties['Other'] = other_parties_count

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_parties, labels=top_parties.index, autopct='%1.1f%%', startangle=140)
plt.title('Party Fragmentation')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```

```python
# 9. High-profile Candidates
high_profile_candidates = input_data[input_data['Leading Candidate'].isin(['NARENDRA MODI', 'RAHUL GANDHI'])]

# Plotting the graph
plt.figure(figsize=(10, 6))
plt.bar(high_profile_candidates['Leading Candidate'], high_profile_candidates['Margin'], color=['#1f77b4', '#ff7f0e'])
plt.title('Votes for Narendra Modi and Rahul Gandhi')
plt.xlabel('Candidate')
plt.ylabel('Votes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 10. Trailing Parties by Votes
trailing_parties_votes = input_data.groupby('Trailing Party')['Margin'].sum().nlargest(5)

# Trailing Parties by Seats
trailing_parties_seats = input_data['Trailing Party'].value_counts().nlargest(5)

# Plotting the graph side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Plot for Trailing Parties by Votes
ax1.bar(trailing_parties_votes.index, trailing_parties_votes.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
ax1.set_title('Top 5 Trailing Parties by Votes')
ax1.set_xlabel('Party')
ax1.set_ylabel('Total Votes')
ax1.set_xticklabels(trailing_parties_votes.index, rotation=45)
ax1.set_ylim(0, trailing_parties_votes.max() * 1.1)

# Plot for Trailing Parties by Seats
ax2.bar(trailing_parties_seats.index, trailing_parties_seats.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
ax2.set_title('Top 5 Trailing Parties by Seats')
ax2.set_xlabel('Party')
ax2.set_ylabel('Number of Seats')
ax2.set_xticklabels(trailing_parties_seats.index, rotation=45)
ax2.set_ylim(0, trailing_parties_seats.max() * 1.1)

plt.tight_layout()
plt.show()
```
